```python
df['rankings'] = df['rankings'].apply(lambda x: re.findall(r'\d+', x))
```

```python
df['rankings'] = df['rankings'].apply(lambda x: list(set(x)))
```

```python
df['rankings'] = df['rankings'].apply(lambda x: [int(number) for number in x])
```

```python
def elem_check(list_1, list_2):
    for elem in list_1:
        if elem not in list_2:
            return False
    return True
```

```python
filtered_df = df[df.apply(lambda row: elem_check(row['rankings'], row['sent_indices']), axis=1)]
```
